### 6. (0.5 points) In R the function `rnorm` can be used to generate data drawn from a normal distribution. Find out which function you can use to generate uniformly distributed data instead. Use it to generate 100 random data points between 0 and 100. Then plot a histogram of the data and draw the mean and median as a vertical line in the graph (Use the `ggplot2` package if you like). Keep in mind: You get a new (i.e. different) data set every time you call the function to generate random data points. ###

```r
library(ggplot2)
# Generating uniformly distributed data
random_values = runif(100, min = 0, max = 1)

data <- data.frame(Values=random_values)
# Create a histogram with ggplot2
ggplot(data, aes(x = Values)) +
  geom_histogram(binwidth = 0.1, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = "Histogram of Random Values", x = "Values", y = "Frequency") +
# Add vertical lines for mean and median
  geom_vline(aes(xintercept = mean(Values), color = "Mean"), linetype = "dashed") +
  geom_vline(aes(xintercept = median(Values), color = "Median"), linetype = "dotted") +
# Add a legend
  scale_color_manual(values = c(Mean = "red", Median = "green")) +
  guides(color = guide_legend(title = NULL))
```
